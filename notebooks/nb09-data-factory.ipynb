# 9. Data Factory

```python
from gml.graph.graph_viz import GraphViz
from gml.graph.data_factory import DataFactory
```

## 9.1 Mitarbeiter mit Matching Cluster

### Ungerichtete Variante

```python
graph = DataFactory().create_graph()
graph.draw(layout=GraphViz.SPIRAL_LAYOUT)
```

Legende:
- MC = Matching Cluster
- EM = Employee (Mitarbeiter)
- I = Identity (U-Nummer)
- F = Firstname (Vorname)
- L = Lastname (Nachname)


### Ungerichtete Variante ohne Identity

```python
graph = DataFactory().create_graph(directed=True, typed=True, add_id=False)
graph.draw()
```

### Gerichtete Variante

```python
graph = DataFactory().create_graph(directed=True)
graph.draw(layout=GraphViz.SPIRAL_LAYOUT)
```

### Gerichtete und typisierte Variante

```python
graph = DataFactory().create_graph(directed=True, typed=True)
graph.draw()
```

### Gerichtete und gewichtete Variante
Bei dieser Variante werden anstelle der Typen Gewichte zwischen 0 und 1 auf der Verbindung eingefhrt.
Die Verbinung zum Identity Knoten hat Gewicht 1, die anderen Gewicht 0.5.

```python
graph = DataFactory().create_graph(directed=True, weighted=True)
graph.draw()
```

## 9.2 Mitarbeiter mit Matching Cluster und Data Collection

### Gerichtete und typisierte Variante

```python
graph = DataFactory().create_graph(directed=True, typed=True, add_dc=True)
graph.draw()
```

Legende:
- DC = Data Collection (Daten Sammlung)
- MC = Matching Cluster
- EM = Employee (Mitarbeiter)
- I = Identity (U-Nummer)
- F = Firstname (Vorname)
- L = Lastname (Nachname)

## 9.3 Mehrere Teilgraphen

### Variante 3 Teilgraphen

```python
graph = DataFactory().create_graph(n=3, directed=True)
graph.draw()
```

### Variante 2 Teilgraphen verbunden mit Root Knoten

```python
graph = DataFactory().create_graph(n=2, connected=True, directed=True)
graph.draw()
```

Legende:
- GR: Graph root (used for connected graphs)
- DR: Data root (used for connected graphs)
- OR: Organisation root (used for connected graphs)
- DC = Data Collection (Daten Sammlung)
- MC = Matching Cluster
- EM = Employee (Mitarbeiter)
- I = Identity (U-Nummer)
- F = Firstname (Vorname)
- L = Lastname (Nachname)

## 9.4 Graph mit vielen Teilgraphen

### Graph mit 100 Teilgraphen

```python
graph = DataFactory().create_graph(n=100, connected=True, directed=True)
graph.print_properties()
```

### Graph mit vielen Teilgraphen

```python
n = 5000
graph = DataFactory().create_graph(n=n)

print("Graph with", n, "subgraphs:")
print("- number of nodes =", len(graph.nodes))
print("- number of edges =", len(graph.edges))
```
