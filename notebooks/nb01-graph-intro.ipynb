```python
from gml.graph.graph_builder import GraphBuilder
```

## 1.1 Undirected Graph
Das Beispiel zeigt einen unterichteten Graphen mit **5 Knoten (Nodes/Vertices)** und **6 Verbindungen (Edges/Links)**.

```python
nodes = ['Alice','David', 'Emma', 'Luca', 'Zora']
edges = [('Alice', 'David'), ('Alice', 'Luca'), ('Alice', 'Zora'), ('David', 'Emma'), ('Emma', 'Luca'), ('Luca', 'Zora')]

graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create()
graph.draw()
```

### Eigenschaften
Die **order** eines Grapen gibt die Anzahl Knoten an. Mit der **size** werden die Anzahl Verbindungen angegeben. Mit **neighbors** werden die Nachbarn eines Konten aufgelistet. Mit dem **degree** eines Konten wird dessen Anzahl Verbindungen angeben.

```python
graph.print_properties()
```

## 1.2 Directed Graph
Das n채chste Beispiel zeigt einen gerichteten Graphen (DiGraph). Die Verbindungen sind jetzt mit einer Richtung gekennzeichnet.

```python
graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create_directed()
graph.draw()
```

###  Eigenschaften
Beim gerichteten Graphen sind **order** und **size** identisch zum vorherigen Beispiel. Die **neighbors** sind jetzt unterschiedlich, da nur noch die ausgehenden Verbindungen gez채hlt werden. So hat Zora zum Beispiel keine Nachbarn mehr, da sie keine ausgehenden Verbindungen aufweist. Beim **degree** kann zus채tzlich zwischen ausgehenden und eingehenden Verbindungen unterschieden werden. Dazu sind die beiden neuen Eigenschaften **in_degree** und **out_degree** verf체gbar.

```python
graph.print_properties()
```

## 1.3 Multigraph
Ein Multigraph erlaubt **mehrere Verbindung zwischen zwei Konten**. Auch hier gibt es eine ungerichtete und eine gerichtete Variante. Das Beispiel zeigt einen gerichteten Graphen mit jeweils zwei Verbindungen zwischen Alice und Zora sowie Alice und David.

```python
nodes = ['Alice','David', 'Emma', 'Luca', 'Zora']
edges = [('Alice', 'David'), ('Alice', 'Luca'), ('Alice', 'Zora'), ('David', 'Emma'), ('Emma', 'Luca'), ('Luca', 'Zora'), ('David', 'Alice'), ('Zora', 'Alice')]

graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create_directed()
graph.draw()
```

### Eigenschaften
Die **order** des Graphen ist immer noch identisch mit den vorherigen Beispielen, da keine neuen Knoten hinzugekommen sind. Entsprechend den zus채tzlichen Verbindungen ist jetzt die **size** gr철sser geworden. Ebenfalls ge채ndert haben die **neighbor** und **degree** Eigenschaften.

```python
graph.print_properties()
```

```python
nodes = ['Alice','David', 'Emma', 'Luca', 'Zora']
edges = [('Alice', 'David', 1), ('Alice', 'Luca', 7), ('Alice', 'Zora', 4), ('David', 'Emma', 9), ('Emma', 'Luca', 5), ('Luca', 'Zora', 2)]

graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create()
graph.draw()
```
