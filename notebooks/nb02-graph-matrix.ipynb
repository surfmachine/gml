# 2. Graph Matrix
Die bisherige Repräsentation von Graphen bestand aus einer Liste von Knoten und einer Liste von Verbindungen. Es gibt aber auch kompaktere Formen, wie ein Graph dargestellt werden kann. Eine davon ist die **Adjacency Matrix**. Dabei bilden die Zeilen und Spalten der Matrix die Knoten ab. Die Werte an den Positionen der Matrix repräsentieren die Verbindungen.

```python
from gml.graph.graph_builder import GraphBuilder
```

## 2.1 Undirected Graph
Das Beispiel zeigt einen unterichteten Graphen mit **5 Knoten (Nodes/Vertices)** und **6 Verbindungen (Edges/Links)**.

```python
nodes = ['Alice','David', 'Emma', 'Luca', 'Zora']
edges = [('Alice', 'David'), ('Alice', 'Luca'), ('Alice', 'Zora'), ('David', 'Emma'), ('Emma', 'Luca'), ('Luca', 'Zora')]

graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create()
graph.draw()
```

### Matrix
Die **Adjacency Matrix** des Graphen zeigt die Verbindungen zwischen den einzelnen Knoten als Matrix. Bei der folgenden Ausgabe sieht man als Spalten- und Zeilentitel jeweils den Knotennamen. An den Matrix Stellen mit 1 besteht eine Verbindung zwischen den jeweiligen Knoten, mit 0 besteht keine Verbindung.

```python
graph.print_matrix()
```

## 2.2 Directed Graph
Das nächste Beispiel zeigt einen gerichteten Graphen (DiGraph). Die Verbindungen sind jetzt mit einer Richtung gekennzeichnet.

```python
graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create_directed()
graph.draw()
```

### Matrix
Bei der **Adjacency Matrix** eines **gerichteten Graphen** markiert die Zeile den Startpunkt und die Spalte den Endpunkt. Auf der Zeile von Alice sind daher jeweils eine 1 bei den Spalten von David, Luca und Zora. Auf der Zeile von Zora hingegen hat es keine 1 in der Spalte von Alice, da die Verbindung nur von Alice zu Zora führt und nicht umgekehrt. Gegenüber dem vorherigen Beispiel enthält diese Matrix also weniger Verbindungen.

```python
graph.print_matrix()
```

## 2.3 Multigraph
Ein Multigraph erlaubt **mehrere Verbindung zwischen zwei Konten**. Auch hier gibt es eine ungerichtete und eine gerichtete Variante. Das Beispiel zeigt einen gerichteten Graphen mit mehreren Verbindungen zwischen einzelnen Knoten.

```python
nodes = ['Alice','David', 'Emma', 'Luca', 'Zora']
edges = [('Alice', 'David'), ('Alice', 'Luca'), ('Alice', 'Zora'), ('Alice', 'Zora'), ('David', 'Emma'), ('Emma', 'Luca'), ('Luca', 'Zora'), ('David', 'Alice'), ('Zora', 'Alice'), ('Emma', 'Luca')]

graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create_directed()
graph.draw()
```

### Matrix
Das Beispiel zeigt die **Adjacency Matrix** der **gerichteten Multigraph** Variante. Alice hat zwei gerichtete Verbindungen zu Zora. Daher steht in der Zeile von Alice und Spalte von Zora eine 2. Das gleiche sieht man bei den beiden Verbindungen von Emma zu Luca. Auch hier hat entsprechende Matrix Element den Wert 2.

```python
graph.print_matrix()
```

## 2.4 Weighted graphs
Es gibt auch Graphen mit **gewichteten** Verbindungen. Auch hier sind ungerichtete, gerichtete und multigrahp Varianten möglich. Das Beispiel zeigt einen unterichteten Graphen mit 5 Knoten und 6 gewichteten Verbindungen.

```python
nodes = ['Alice','David', 'Emma', 'Luca', 'Zora']
edges = [('Alice', 'David', 1), ('Alice', 'Luca', 7), ('Alice', 'Zora', 4), ('David', 'Emma', 9), ('Emma', 'Luca', 5), ('Luca', 'Zora', 2)]

graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create()
graph.draw()
```

### Matrix
Die **Adjacency Matrix** eines **gewichteten Graphen** beinhaltet nun die Gewichtung als Matrix Eintrag. Der Wert 0 bedeutet nach wie vor keine Verbindung. Werte grösser oder gleich 1 bedeuten eine Verbindung mit der entsprechenden Gewichtung.

```python
graph.print_matrix()
```

## 2.5 Matrix directed and weighted
Ein gewichteter Graph kann natürchlich zusätzlich auch gerichted sein. In der **Adjacency Matrix** erscheinen die Gewichte dann nur an den entsprechenden Stellen, welche die Richtung einer Verbindung repräsentieren.

```python
graph = GraphBuilder().append_nodes(nodes).append_edges(edges).create_directed()
graph.draw()
```

```python
graph.print_matrix()
```
