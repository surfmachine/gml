# 5. Graph Machine Learning Embedding

Beim Embedding geht es darum, mit einer **Abbildungsfunktion** einem diskreten **Graphen** zu einer niedrigdimensionale **Vektordarstellung** zu transormieren. Ziel ist, dass die lokalen und globalen Eigenschaften des Graphen möglichst erhalten bleiben. Daraus ergibt sich, dass Graphen, Knoten oder Verbindungen, die im ursprünglichen Raum ähnlich sind, auch im neuen Raum ähnlich sein werden.

Nachfolgend werden je ein Besipiel zum Node-, Edge- und Graph-Embedding gezeigt. Bei den vorgestellten Methoden (Node2Vec, Edge2Vec, und Graph2Vec) handelt es sich um sogenannte **Shallow Embedding** Methoden. Diese "flachen" Einbettungsmethoden geben ein eine vektorielle Repräsentation der gelieferten Daten zurück. Es ist nicht möglich den Einbettungsvektor für ungesehene Daten zu erhalten.

```python
from gml.graph.graph_builder import GraphBuilder
from gml.graph.graph_generator import GraphGenerator
from gml.graph.graph_embedding import GraphEmbedding
```

## 5.1 Node embedding

Das folgende Beispiel zeigt die Verwendung der **Node2Vec** Abbildungsfunktion (Details siehe Modul graph_embedding). Dabei werden die **Knoten** des folgenden Graphen in einen **2 dimensonalen Vektorraum** transformiert. Wie man sieht sind die eng miteinander verknüpften Knoten 1-4, 6-8 und 10-12 auch in der Scatterplot Darstellung nahe beieinander, während dem die anderen Konten weiter weg sind. Dies liegt daran, dass der Algorithmus u.a. die Nachbarschaftsinformationen jedes Knotens verwendet, um die Darstellung zu erzeugen. Dies ermöglicht eine klare Unterscheidung der genannten Gruppen.

```python
edges = [(1,2),(1,3),(1,4),(2,3),(2,4),(3,4),(4,5),(5,6),(6,7),(6,8),(7,8),(8,9),(9,10),(10,11),(10,12), (11,12)]
graph = GraphBuilder().append_edges(edges).create()
graph.draw()

graph.create_node_embedding().draw()
```

## 5.2 Edge embedding

In diesem Beispiel wird die Verwendung der **Edge2Vec** Abbildungsfunktion (Details siehe Modul graph_embedding) gezeigt. Hier werden die **Verbindungen** des Graphen in einen **2 dimensonalen Vektorraum** transformiert. Auch hier werden die nahe beieinander liegenden Verbindungen in der Scatterplot Darstellung entsprechend gruppiert dargestellt. Die Verbindungsinformationen sind also auch im neuen Vektorraum ersichtlich.

```python
graph.draw()

graph.create_edge_embedding().draw()
```

## 5.3 Graph embedding

Das dritte Beispiel zeigt die Anwendung des **Grap2Vec** Algorithmus zur Transformation von Graphen in Vektoren (Details siehe wiederum im Modul graph_embedding). Als Graphen werden 12 zufällig erzeugte Watts Stragatz Graphen mit unterschiedlichen Eigenschafen erzeugt. Die erzeugten Graphen werden anchliessend wieder in einen **2 dimensonalen Vektorraum** transformiert. Auch hier ist ersichtlich, dass Graphen mit ähnlichen Eigenschaften (Distanzen) im Vektorraum näher beieinander liegen als Graphen mit unterschiedlichen Eigenschaften.

_Beachte: Eine explizite Referenzierung der Graph Nummern ist in diesem Besipiel nicht möglich, da in jemden Durchgang die zufällig generierten Graphen anders sind._

```python
n_graphs = 12
graphs = GraphGenerator().append_random_watts_stragatz(n_graphs).create()

for i, g in enumerate(graphs):
    g.draw(title=f"Graph {i}")

GraphEmbedding(graphs).draw()
```
