# 3. Graph Edge List
Eine weitere Möglichkeit für die Repräsentation von Graphen ist die sogenannte **Edge List**. Dabei werden die Verbindungen als Liste geführt. Die Knoten können anhand der Verbindungsangaben automatisch detektiert werden und müssen nicht explizit angegeben werden.

```python
from gml.graph.graph_builder import GraphBuilder
```

## Undirected Graph
Das Beispiel zeigt einen unterichteten Graphen mit **5 Knoten (Nodes/Vertices)** und **6 Verbindungen (Edges/Links)**.

```python
edges = [('Alice', 'David'), ('Alice', 'Luca'), ('Alice', 'Zora'), ('David', 'Emma'), ('Emma', 'Luca'), ('Luca', 'Zora')]

graph = GraphBuilder().append_edges(edges).create()
graph.draw()
```

### Edge List
Die **Edge List** zeigt alle Verbindungen der Graphen. Bei einem ungerichteten Graphen spielt die Richtung keine Rolle. Es wird also nicht zwischen Quelle und Ziel unterschieden.

```python
graph.print_edgelist()
```

## Directed Graph
Das nächste Beispiel zeigt einen gerichteten Graphen (DiGraph). Die Verbindungen sind jetzt mit einer Richtung gekennzeichnet.

```python
graph = GraphBuilder().append_edges(edges).create_directed()
graph.draw()
```

### Edge List
Entsprechend der gerichteten Verbindungen, zeigt nun auch die **Edge List** die Quelle (Source) und das Ziel (Target) explizit an.

```python
graph.print_edgelist()
```

## Multigraph
Ein Multigraph erlaubt **mehrere Verbindung zwischen zwei Konten**. Auch hier gibt es eine ungerichtete und eine gerichtete Variante. Das Beispiel zeigt einen gerichteten Graphen mit jeweils zwei Verbindungen zwischen Alice und Zora sowie Alice und David.

```python
edges = [('Alice', 'David'), ('Alice', 'Luca'), ('Alice', 'Zora'), ('David', 'Emma'), ('Emma', 'Luca'), ('Luca', 'Zora'), ('David', 'Alice'), ('Zora', 'Alice')]

graph = GraphBuilder().append_edges(edges).create_directed()
graph.draw()
```

### Edge List
Das Beispiel zeigt die **Edge List** der **gerichteten Multigraph** Variante. Im Vergleich zur vorherigen Varianten werden die weiteren Verbindungen in der Liste zusätzlich aufgeführt.

```python
graph.print_edgelist()
```

## Weighted graphs
Es gibt auch Graphen mit **gewichteten** Verbindungen. Auch hier sind ungerichtete, gerichtete und multigrahp Varianten möglich. Das Beispiel zeigt einen unterichteten Graphen mit 5 Knoten und 6 gewichteten Verbindungen.

```python
edges = [('Alice', 'David', 1), ('Alice', 'Luca', 7), ('Alice', 'Zora', 4), ('David', 'Emma', 9), ('Emma', 'Luca', 5), ('Luca', 'Zora', 2)]

graph = GraphBuilder().append_edges(edges).create()
graph.draw()
```

### Edge List
Die **Edge List** des **gewichteten Graphen** beinhaltet neben den beiden beteiligten Knoten nun auch die Angabe der Gewichtung.

```python
graph.print_edgelist()
```

## Multiple properties
Einer Verbindung können beliebige Eigenschaften zugeordnet werden. Im folgenden Beispiel hat jede Verbindung eine Angabe zur Distanz (Dist) und den Kosten (Cost).

```python
edges = [
    ('Alice', 'Zora', {'Dist':1, 'Cost':3}),
    ('Alice', 'Luca', {'Dist':5, 'Cost':2}),
    ('Luca', 'Zora', {'Dist':2, 'Cost':1})
]
graph = GraphBuilder().append_edges(edges).create()
graph.draw()
```

### Edge List
Die **Edge List** zum obigen Graphen, zeigt nun auch alle zugewiesenen Eigenschaften der Verbindungen.

```python
graph.print_edgelist()
```
